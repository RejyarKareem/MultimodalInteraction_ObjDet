```python
from PIL import ImageDraw
from IPython.display import display

def plot_bounding_boxes(im, boxes, labels=None, normalized=True):
    """
    Draw bounding boxes on a PIL image and display it in a Jupyter notebook.

    Parameters
    ----------
    im : PIL.Image
        The input image.

    boxes : list of list[float]
        List of bounding boxes in the format:
        [x_min, y_min, x_max, y_max].

    labels : list[str], optional
        Optional list of labels for each bounding box. The label will be
        drawn inside a colored rectangle for readability.

    normalized : bool, default=True
        If True, box coordinates are assumed to be normalized in the
        range [0,1000] and will be converted to pixel coordinates using
        the image size.

    Returns
    -------
    None
        Displays the image with bounding boxes directly in the notebook.
    """
    img = im.copy()
    draw = ImageDraw.Draw(img)
    colors = ["red", "green", "blue", "yellow", "orange"]

    w, h = img.size

    for i, (x_min, y_min, x_max, y_max) in enumerate(boxes):
        if normalized:
            x_min = x_min / 1000 * w
            x_max = x_max / 1000 * w
            y_min = y_min / 1000 * h
            y_max = y_max / 1000 * h

        color = colors[i % len(colors)]

        draw.rectangle(
            [(x_min, y_min), (x_max, y_max)],
            outline=color,
            width=4,
        )

        if labels:
            text = labels[i]

            # draw background box for readability
            bbox = draw.textbbox((x_min, y_min), text)
            draw.rectangle(bbox, fill=color)

            draw.text((x_min, y_min), text, fill="white")

    display(img)
```
